```python
import numpy as np
import pandas as pd
from Binary_Logistic_Regression import BinaryLogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('tumor.csv')
```

```python
data
```

```python
data.drop(labels=[data.columns[0],data.columns[-1]], axis=1, inplace=True)
```

```python
data
```

```python
class BinaryLR_Tumor(BinaryLogisticRegression):
    def __init__(self, data, n_comp, LR=10**(-6), ET=10**(-5)):
        d = data.copy()
        labels = d['diagnosis']
        d = np.array(d.iloc[:,1:])
        
        pca = PCA(n_comp)
        d = pca.fit_transform(d)
        d = pd.DataFrame(d)
        d['diagnosis'] = labels
        d['diagnosis'].replace(to_replace=['M','B'], value=[1,0],inplace=True)
        
        data_m = pd.DataFrame(d[d['diagnosis'] == 1])
        data_b = pd.DataFrame(d[d['diagnosis'] == 0])
        
        self.train_data = pd.concat([data_m.iloc[:199,:],data_b.iloc[:199,:]], axis=0)
        
        remaining_data = pd.concat([data_m.iloc[199:,:],data_b.iloc[199:,:]], axis=0)
        
        X_cv, X_test, Y_cv, Y_test = train_test_split(remaining_data.iloc[:,:-1], remaining_data['diagnosis'], test_size=1/3)
        
        self.cv_data = pd.DataFrame(X_cv)
        self.cv_data['diagnosis'] = Y_cv
        
        self.test_data = pd.DataFrame(X_test)
        self.test_data['diagnosis'] = Y_test
        
        del data_m, data_b, remaining_data, X_cv, X_test, Y_cv, Y_test, d, labels, pca
        
        super().__init__(1, LearningRate=LR, ErrorTolerance=ET)
```

```python
def evaluate(predicted, actual):
    TP = np.count_nonzero((predicted == 1) & (actual == 1))
    TN = np.count_nonzero((predicted == 0) & (actual == 0))
    FP = np.count_nonzero((predicted == 1) & (actual == 0))
    FN = np.count_nonzero((predicted == 0) & (actual == 1))
    
    
    if (TP + TN + FP + FN) == 0:
        accuracy = 0
    else:
        accuracy = (TP + TN)/(TP + TN + FP + FN)
    
    if (TP + FP) == 0:
        precision = 0
    else:
        precision = TP/(TP + FP)
    
    if (TP + FN) == 0:
        recall = 0
    else:
        recall = TP/(TP + FN)
        
    if (precision + recall) == 0:
        f1_score = 0
    else:
        f1_score = (2 * precision * recall)/(precision + recall)
    
    return (accuracy, precision, recall, f1_score)
```

```python
Results = dict()
best = {'Accuracy':0, 'Precision':0, 'Recall':0, 'F1 Score':0, 'Best Obj':'obj', 'Parameter':()}
```

# Trying Different Combinaitions of Hyperparameters

```python
for n_comp in range(2,18):
    for LR in [5,6,7,8,9,10]:
        for ET in [5,6,7,8,9,10]:
            obj = BinaryLR_Tumor(data, n_comp, LR=10**(-LR), ET=10**(-ET))
            obj.fit(obj.train_data.iloc[:,:-1], obj.train_data['diagnosis'])
            obj.predict(obj.cv_data.iloc[:,:-1])
            Results[(n_comp, LR, ET)] = evaluate(obj.predicted_labels, obj.cv_data['diagnosis'])
            if (Results[(n_comp, LR, ET)][0] + Results[(n_comp, LR, ET)][1] + Results[(n_comp, LR, ET)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
                best['Accuracy'] = Results[(n_comp, LR, ET)][0]
                best['Precision'] = Results[(n_comp, LR, ET)][1]
                best['Recall'] = Results[(n_comp, LR, ET)][2]
                best['F1 Score'] = Results[(n_comp, LR, ET)][3]
                best['Best Obj'] = obj
                best['Parameter'] = (n_comp, LR, ET)
            else:
                del obj
```

# From over 612 Hyperparameter combinaitons we find our best model
