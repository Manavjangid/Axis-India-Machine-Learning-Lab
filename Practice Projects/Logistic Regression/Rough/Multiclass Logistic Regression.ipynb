```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
train_path = ".\DevanagariHandwrittenCharacterDataset\Train"

test_path = ".\DevanagariHandwrittenCharacterDataset\Test"
```

```python
def data(path):
    df = list()
    labels = list()
    for image_folder in [os.path.join(path, folder_name) for folder_name in os.listdir(path)]:
        for image_path in [os.path.join(image_folder, image) for image in os.listdir(image_folder)]:
            labels.append(image_folder.split('\\')[-1])
            image = plt.imread(image_path)
            df.append(np.reshape(image, (1,image.shape[0] * image.shape[1])))
    df = np.concatenate(df,axis=0)
    df = pd.DataFrame(df)
    df['labels'] = labels
    return df
```

```python
def one_hot_encoding(df):
    data_array = list()
    column_name = list()
    df = pd.DataFrame(df)
    for column in df.columns:
        unique = df[column].unique()
        size = len(df[column].unique())
        df[column].replace(to_replace=unique, value=range(size), inplace=True)
        data_array.append(np.eye(size, size)[df[column]])
        for u in unique:
            column_name.append(column + '_' + str(u))
    return pd.DataFrame(np.concatenate(data_array, axis=1), columns=column_name)
```

```python
train_data = data(train_path)
test_data = data(test_path)
```

```python
train = train_data.iloc[:,:-1]
train_labels = train_data['labels']
test = test_data.iloc[:,:-1]
test_labels = test_data['labels']

train = (train - train.mean(axis=0))/train.std(axis=0)
test = (test - test.mean(axis=0))/test.std(axis=0)

train.replace(to_replace=np.nan, value=0, inplace=True)
test.replace(to_replace=np.nan, value=0, inplace=True)
```

```python
Q = eigen_vector_calculation()
```

```python
train = np.matmul(np.array(train),Q)
test = np.matmul(np.array(test),Q)

train = pd.DataFrame(train)
train['labels'] = train_labels

test = pd.DataFrame(test)
test['labels'] = test_labels
```

```python
labels = one_hot_encoding((train.iloc[:,-1]))
train.drop(['labels'], axis=1, inplace=True)
train = pd.concat([train, labels], axis=1)
```

```python
def softmax_func(theta_hat_0, theta_hat, x, y):
    x = np.array(x)
    y = np.array(y)
    
    norm_p = np.exp(-np.matmul(np.matmul(x, theta_hat.T), theta_hat_0))
    
    numerator_p = np.exp(-np.matmul(x, theta_hat.T)) + theta_hat_0
    
    posterior = numerator_p/np.reshape(norm_p, (norm_p.shape[0],-1))
    
    posterior = np.log(posterior)
    
    posterior = np.array(pd.DataFrame(posterior).replace(to_replace=[np.nan,np.inf,np.NINF], value=[0,0,0]))
    
    posterior = posterior * y
    
    return -np.sum(posterior)
```
