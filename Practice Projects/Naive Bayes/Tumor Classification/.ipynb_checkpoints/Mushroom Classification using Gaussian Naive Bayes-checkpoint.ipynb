```python
import numpy as np
import pandas as pd
from Gaussian_Naive_Bayes import GaussianNaiveBayes
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('data.csv')
```

```python
data
```

```python
data.drop(labels=[data.columns[0],data.columns[-1]], axis=1, inplace=True)
```

```python
data
```

```python
class GaussianNB_Mushroom(GaussianNaiveBayes):
    def __init__(self, data, n_comp, da='qda', rda_p=np.nan):
        d = data.copy()
        labels = d['diagnosis']
        d = np.array(d.iloc[:,1:])
        
        pca = PCA(n_comp)
        d = pca.fit_transform(d)
        d = pd.DataFrame(d)
        d['diagnosis'] = labels
        
        data_m = pd.DataFrame(d[d['diagnosis'] == 'M'])
        data_b = pd.DataFrame(d[d['diagnosis'] == 'B'])
        
        self.train_data = pd.concat([data_m.iloc[:199,:],data_b.iloc[:199,:]], axis=0)
        
        remaining_data = pd.concat([data_m.iloc[199:,:],data_b.iloc[199:,:]], axis=0)
        
        X_cv, X_test, Y_cv, Y_test = train_test_split(remaining_data.iloc[:,:-1], remaining_data['diagnosis'], test_size=1/3)
        
        self.cv_data = pd.DataFrame(X_cv)
        self.cv_data['diagnosis'] = Y_cv
        
        self.test_data = pd.DataFrame(X_test)
        self.test_data['diagnosis'] = Y_test
        
        del data_m, data_b, remaining_data, X_cv, X_test, Y_cv, Y_test, d, labels, pca
        
        super().__init__(da,rda_p)
```

```python
def evaluate(predicted, actual):
    TP = np.count_nonzero((predicted == 'M') & (actual == 'M'))
    TN = np.count_nonzero((predicted == 'B') & (actual == 'B'))
    FP = np.count_nonzero((predicted == 'M') & (actual == 'B'))
    FN = np.count_nonzero((predicted == 'B') & (actual == 'M'))
    
    accuracy = (TP + TN)/(TP + TN + FP + FN)
    
    if (TP + FP) == 0:
        precision = 0
    else:
        precision = TP/(TP + FP)
    
    if (TP + FN) == 0:
        recall = 0
    else:
        recall = TP/(TP + FN)
        
    if (precision + recall) == 0:
        f1_score = 0
    else:
        f1_score = (2 * precision * recall)/(precision + recall)
    
    return (accuracy, precision, recall, f1_score)
```

```python
Results = dict()
best = {'Accuracy':0, 'Precision':0, 'Recall':0, 'F1 Score':0, 'Best Obj':'obj', 'Parameter':()}
```

# We will check for various values of N Components in PCA for assumption of Quadratic Discriminant Analysis

```python
for i in range(2,18): 
    obj = GaussianNB_Mushroom(data, i, 'qda')
    obj.fit(obj.train_data.iloc[:,:-1],obj.train_data['diagnosis'])
    obj.predict(obj.cv_data.iloc[:,:-1])
    Results[('qda', i)] = evaluate(obj.predicted_labels, obj.cv_data['diagnosis'])
    if (Results[('qda', i)][0] + Results[('qda', i)][1] + Results[('qda', i)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
        best['Accuracy'] = Results[('qda', i)][0]
        best['Precision'] = Results[('qda', i)][1]
        best['Recall'] = Results[('qda', i)][2]
        best['F1 Score'] = Results[('qda', i)][3]
        best['Best Obj'] = obj
        best['Parameter'] = ('qda', i)
    else:
        del obj
```

```python
# best model so far according to accuracy
best
```

# We will check for various values of N Components in PCA for assumption of Linear Discriminant Analysis

```python
for i in range(2,18): 
    obj = GaussianNB_Mushroom(data, i, 'lda')
    obj.fit(obj.train_data.iloc[:,:-1],obj.train_data['diagnosis'])
    obj.predict(obj.cv_data.iloc[:,:-1])
    Results[('lda', i)] = evaluate(obj.predicted_labels, obj.cv_data['diagnosis'])
    if (Results[('lda', i)][0] + Results[('lda', i)][1] + Results[('lda', i)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
        best['Accuracy'] = Results[('lda', i)][0]
        best['Precision'] = Results[('lda', i)][1]
        best['Recall'] = Results[('lda', i)][2]
        best['F1 Score'] = Results[('lda', i)][3]
        best['Best Obj'] = obj
        best['Parameter'] = ('lda', i)
    else:
        del obj
```

```python
best
```

# We will check for various values of N Components in PCA for assumption of Regularized Discriminant Analysis

```python
for alpha in np.arange(0.1,1,0.1):
    for gamma in np.arange(0.1,1,0.1):
        for i in range(2,18): 
            obj = GaussianNB_Mushroom(data, i, 'rda', (alpha, gamma))
            obj.fit(obj.train_data.iloc[:,:-1],obj.train_data['diagnosis'])
            obj.predict(obj.cv_data.iloc[:,:-1])
            Results[('rda', i, alpha, gamma)] = evaluate(obj.predicted_labels, obj.cv_data['diagnosis'])
            if (Results[('rda', i, alpha, gamma)][0] + Results[('rda', i, alpha, gamma)][1] + Results[('rda', i, alpha, gamma)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
                best['Accuracy'] = Results[('rda', i, alpha, gamma)][0]
                best['Precision'] = Results[('rda', i, alpha, gamma)][1]
                best['Recall'] = Results[('rda', i, alpha, gamma)][2]
                best['F1 Score'] = Results[('rda', i, alpha, gamma)][3]
                best['Best Obj'] = obj
                best['Parameter'] = ('rda', i, alpha, gamma)
            else:
                del obj
```

```python
best
```

# From over 1328 Hyperparameter combinaitons we find our best model

```python
best_obj = best['Best Obj']
```

```python
best_obj.predict(best_obj.test_data.iloc[:,:-1])
```
