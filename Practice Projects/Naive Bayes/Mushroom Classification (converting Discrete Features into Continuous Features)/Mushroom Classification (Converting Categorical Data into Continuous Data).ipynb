# In this we will convert categorical data into continuous data, which will definetly improve model performance.

```python
import numpy as np
import pandas as pd
from Gaussian_Naive_Bayes import GaussianNaiveBayes
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('mushrooms.csv')
```

```python
data
```

```python
def one_hot_encoding(df):
    data_array = list()
    column_name = list()
    for column in df.columns:
        unique = df[column].unique()
        size = len(df[column].unique())
        df[column].replace(to_replace=unique, value=range(size), inplace=True)
        data_array.append(np.eye(size, size)[df[column]])
        for u in unique:
            column_name.append(column + '_' + str(u))
    return pd.DataFrame(np.concatenate(data_array, axis=1), columns=column_name)
```

```python
new_data = one_hot_encoding(data.iloc[:,1:])
```

```python
new_data = (new_data - new_data.mean(axis=0))/new_data.std(axis=0)
```

```python
new_data.dropna(axis=1, inplace=True)
```

```python
new_data
```

```python
class GNB_Mushroom(GaussianNaiveBayes):
    def __init__(self,X,Y,n_comp,discriminent_analysis_assumption='qda',rda_p=np.nan):
        
        pca = PCA(n_components=n_comp)
        self.data_after_pca = pd.DataFrame(pca.fit_transform(X))
        
        self.data_after_pca['class'] = Y
        
        p = self.data_after_pca[self.data_after_pca['class'] == 'p']
        e = self.data_after_pca[self.data_after_pca['class'] == 'e']
        
        self.train = pd.concat([p.iloc[:int(X.shape[0] * 0.7)//2,:], e.iloc[:int(X.shape[0] * 0.7)//2,:]], axis=0)
        
        remaining = pd.concat([p.iloc[int(X.shape[0] * 0.7)//2:,:], e.iloc[int(X.shape[0] * 0.7)//2:,:]])
        
        X_cv, X_test, Y_cv, Y_test = train_test_split(remaining.iloc[:,:-1],remaining['class'],test_size=1/2)
        
        self.cv = pd.DataFrame(X_cv)
        self.cv['class'] = Y_cv
        
        self.test = pd.DataFrame(X_test)
        self.test['class'] = Y_test
        
        del pca, p, e, remaining, X_cv, X_test, Y_cv, Y_test
        
        super().__init__(discriminent_analysis_assumption, rda_p)
```

```python
def evaluate(predicted, actual):
    TP = np.count_nonzero((predicted == 'p') & (actual == 'p'))
    TN = np.count_nonzero((predicted == 'e') & (actual == 'e'))
    FP = np.count_nonzero((predicted == 'p') & (actual == 'e'))
    FN = np.count_nonzero((predicted == 'e') & (actual == 'p'))
    
    accuracy = (TP + TN)/(TP + TN + FP + FN)
    
    if (TP + FP) == 0:
        precision = 0
    else:
        precision = TP/(TP + FP)
    
    if (TP + FN) == 0:
        recall = 0
    else:
        recall = TP/(TP + FN)
        
    if (precision + recall) == 0:
        f1_score = 0
    else:
        f1_score = (2 * precision * recall)/(precision + recall)
    
    return (accuracy, precision, recall, f1_score)
```

```python
Results = dict()
best = {'Accuracy':0, 'Precision':0, 'Recall':0, 'F1 Score':0, 'Best Obj':'obj', 'Parameter':()}
```

# We will check for various values of N Components in PCA for assumption of Quadratic Discriminant Analysis

```python
for i in range(2,36): 
    obj = GNB_Mushroom(new_data, data['class'], i, 'qda')
    obj.fit(obj.train.iloc[:,:-1],obj.train['class'])
    obj.predict(obj.cv.iloc[:,:-1])
    Results[('qda', i)] = evaluate(obj.predicted_labels, obj.cv['class'])
    if (Results[('qda', i)][0] + Results[('qda', i)][1] + Results[('qda', i)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
        best['Accuracy'] = Results[('qda', i)][0]
        best['Precision'] = Results[('qda', i)][1]
        best['Recall'] = Results[('qda', i)][2]
        best['F1 Score'] = Results[('qda', i)][3]
        best['Best Obj'] = obj
        best['Parameter'] = ('qda', i)
    else:
        del obj
```

```python
best
```

# We will check for various values of N Components in PCA for assumption of Linear Discriminant Analysis

```python
for i in range(2,60): 
    obj = GNB_Mushroom(new_data, data['class'], i, 'lda')
    obj.fit(obj.train.iloc[:,:-1],obj.train['class'])
    obj.predict(obj.cv.iloc[:,:-1])
    Results[('lda', i)] = evaluate(obj.predicted_labels, obj.cv['class'])
    if (Results[('lda', i)][0] + Results[('lda', i)][1] + Results[('lda', i)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
        best['Accuracy'] = Results[('lda', i)][0]
        best['Precision'] = Results[('lda', i)][1]
        best['Recall'] = Results[('lda', i)][2]
        best['F1 Score'] = Results[('lda', i)][3]
        best['Best Obj'] = obj
        best['Parameter'] = ('lda', i)
    else:
        del obj
```

```python
best
```

# We will check for various values of N Components in PCA for assumption of Regularized Discriminant Analysis

```python
for alpha in np.arange(0.1,1,0.1):
    for gamma in np.arange(0.1,1,0.1):
        for i in range(2,117):
            obj = GNB_Mushroom(new_data, data['class'], i, 'rda', (alpha, gamma))
            obj.fit(obj.train.iloc[:,:-1],obj.train['class'])
            obj.predict(obj.cv.iloc[:,:-1])
            Results[('rda', i, alpha, gamma)] = evaluate(obj.predicted_labels, obj.cv['class'])
            if (Results[('rda', i, alpha, gamma)][0] + Results[('rda', i, alpha, gamma)][1] + Results[('rda', i, alpha, gamma)][2]) > (best['Accuracy'] + best['Precision'] + best['Recall']):
                best['Accuracy'] = Results[('rda', i, alpha, gamma)][0]
                best['Precision'] = Results[('rda', i, alpha, gamma)][1]
                best['Recall'] = Results[('rda', i, alpha, gamma)][2]
                best['F1 Score'] = Results[('rda', i, alpha, gamma)][3]
                best['Best Obj'] = obj
                best['Parameter'] = ('rda', i, alpha, gamma)
            else:
                del obj
```

```python
best
```
