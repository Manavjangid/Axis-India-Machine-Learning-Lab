\begin{equation}
P(Cap-shape \cap Cap-surface | Class = p)\\[10pt]
\end{equation}

```python
class_p=data[data['class']=='p']
class_p
```

```python
mask=(class_p['cap-shape']=='p') & (class_p['cap-surface']=='p')
Numerator=class_p[mask]
```

```python
Denominator=class_p
```

```python
P=Numerator.shape[0]/Denominator.shape[0]
P
```

\begin{equation}
P(Cap-shape \cap Cap-surface \cap ......\cap habitat | Class = p)\\[10pt]
\end{equation}

```python
probabilities_p = {}
denominator = data[data['class'] == 'p'].shape[0]
```

```python
for column_name in data.columns[1:]:
    probabilities_p[column_name] = {}
    for unique_value in data[column_name].unique():
        probabilities_p[column_name][unique_value] = (np.count_nonzero(data[data['class'] == 'p'][column_name] == unique_value)/denominator)
```

```python
# Here are all the Probabilities there Combination will result in Answer of the Given probability
probabilities_p
```
