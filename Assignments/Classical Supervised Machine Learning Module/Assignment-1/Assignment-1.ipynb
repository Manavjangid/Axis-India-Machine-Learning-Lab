# Now we will Apply PCA:-

```python
def pca(data, n_components):
    lables = data['diagnosis']
    X = np.array(data.iloc[:,1:])
    X_dash = X - X.mean(axis=0)
    cov = (1/X.shape[0]) * np.matmul(X_dash.T, X_dash)
    Q = np.linalg.svd(cov)[0]
    Q_tilda = Q[:,:n_components]
    X_new = np.matmul(X_dash, Q_tilda)
    data_new = pd.DataFrame(X_new)
    data_new['diagnosis'] = lables
    return data_new
```

# Fitting Gaussian Naive Bayes on Data i.e Calculating Values of Parameters:-

```python
def fit(data):
    training_data_m = data[data['diagnosis'] == 'M'].iloc[:,:-1]
    training_data_b = data[data['diagnosis'] == 'B'].iloc[:,:-1]
    mean = np.array([training_data_m.mean(axis=0), training_data_b.mean(axis=0)])
    cov = np.array([training_data_m.cov(), training_data_b.cov()])
    prior = np.array([training_data_m.shape[0]/data.shape[0], training_data_b.shape[0]/data.shape[0]])
    return mean, cov, prior
```

```python
mean, covariance_matrix, prior_p= fit(Training_data)
```

# Evaluation Function:-

```python
def evaluate(data):
    posterior_m = np.array(s.multivariate_normal.pdf(data.iloc[:,:-1], mean[0], covariance_matrix[0]) * prior_p[0])
    posterior_b = np.array(s.multivariate_normal.pdf(data.iloc[:,:-1], mean[1], covariance_matrix[1]) * prior_p[1])
    Boolean_mask = pd.Series(posterior_b > posterior_m)
    return np.array(Boolean_mask.replace(to_replace=[True, False], value=['B', 'M']))
```

# Performance Matric Calculation:-

```python
#This Function will only work with Biclassification.
def performance(actual,predicted,positive):
    unique = np.unique(actual)
    
    if positive == unique[0]:
        negative = unique[1]
    else:
        negative = unique[0]
        
    TP = np.count_nonzero((predicted == positive) & (actual == positive))
    TN = np.count_nonzero((predicted == negative) & (actual == negative))
    FP = np.count_nonzero((predicted == positive) & (actual == negative))
    FN = np.count_nonzero((predicted == negative) & (actual == positive))
    
    if (TP+TN+FP+FN) == 0:
        Accuracy = 0
    else:
        Accuracy = (TP+TN)/(TP+TN+FP+FN)
    
    if (TP+FP) == 0:
        Precision = 0
    else:
        Precision = TP/(TP+FP)
    
    if (TP+FN) == 0:
        Recall = 0
    else:
        Recall = TP/(TP+FN)
    
    if (Precision+Recall) == 0:
        F1_Score = 0
    else:
        F1_Score = (2*Precision*Recall)/(Precision+Recall)
    
    print(f' Accuracy = {Accuracy} \n Precision = {Precision} \n Recall = {Recall} \n F1-Score = {F1_Score}')
    
    return {'Accuracy':Accuracy, 'Precision':Precision, 'Recall':Recall, 'F1-Score':F1_Score}
```
